```python

# Import necessary modules for diff metrics
import sys
import os
sys.path.append(os.path.abspath(os.path.join('..')))
from utils.simple_metrics import diff_score
import random
def evaluate_diff_metrics(subset_size=100, traceback_error_type=None, random_seed=42):
    """
    Evaluate diff metrics on a subset of the buggy training dataset.
    
    Args:
        subset_size (int): Number of samples to evaluate (default: 100)
        traceback_error_type (str, optional): Specific error type to filter by (default: None)
        random_seed (int): Random seed for reproducibility (default: 42)
    
    Returns:
        dict: Dictionary containing results and statistics
    """
    
    # Set random seed for reproducibility
    random.seed(random_seed)
    
    # Filter dataset based on traceback error type if provided
    if traceback_error_type:
        filtered_df = buggy_train_df[buggy_train_df['traceback_type'] == traceback_error_type].copy()
        print(f"Filtered dataset to {len(filtered_df)} samples with error type: {traceback_error_type}")
        
        if len(filtered_df) < subset_size:
            print(f"Warning: Only {len(filtered_df)} samples available for error type '{traceback_error_type}'")
            subset_size = len(filtered_df)
    else:
        filtered_df = buggy_train_df.copy()
        print(f"Using all {len(filtered_df)} samples (no error type filter)")
    
    # Create a subset of random samples
    subset_indices = random.sample(range(len(filtered_df)), min(subset_size, len(filtered_df)))
    subset_df = filtered_df.iloc[subset_indices].copy()
    
    print(f"Created subset of {len(subset_df)} samples for diff metric evaluation")
    
    # Prepare data for evaluation
    before_codes = subset_df['before_merge'].tolist()
    after_codes = subset_df['after_merge'].tolist()
    
    # Run diff metric evaluation on the subset
    print("Running diff metric evaluation...")
    results = []
    
    for i, (before, after) in enumerate(zip(before_codes, after_codes)):
        result = diff_score(before, after)
        results.append(result)
    
    # Calculate aggregate statistics for both AST and text scores
    ast_scores = [r['ast_score'] for r in results]
    text_scores = [r['text_score'] for r in results]
    
    avg_ast_score = sum(ast_scores) / len(ast_scores)
    avg_text_score = sum(text_scores) / len(text_scores)
    
    print(f"\nDiff Metric Results:")
    print(f"AST Score - Average: {avg_ast_score:.4f}, Min: {min(ast_scores):.4f}, Max: {max(ast_scores):.4f}")
    print(f"Text Score - Average: {avg_text_score:.4f}, Min: {min(text_scores):.4f}, Max: {max(text_scores):.4f}")
    
    # Show distribution of both scores
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
    
    # AST scores histogram
    ax1.hist(ast_scores, bins=20, alpha=0.7, edgecolor='black')
    ax1.set_xlabel('AST Similarity Score')
    ax1.set_ylabel('Frequency')
    ax1.set_title(f'Distribution of AST Similarity Scores ({len(subset_df)} samples)')
    ax1.grid(alpha=0.3)
    
    # Text scores histogram
    ax2.hist(text_scores, bins=20, alpha=0.7, edgecolor='black')
    ax2.set_xlabel('Text Similarity Score')
    ax2.set_ylabel('Frequency')
    ax2.set_title(f'Distribution of Text Similarity Scores ({len(subset_df)} samples)')
    ax2.grid(alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    # Show some example results
    print(f"\nSample Results (first 5):")
    for i, result in enumerate(results[:5]):
        print(f"Sample {i+1}: AST Score = {result['ast_score']:.4f}, Text Score = {result['text_score']:.4f}")
    
    # Return results dictionary
    return {
        'results': results,
        'ast_scores': ast_scores,
        'text_scores': text_scores,
        'avg_ast_score': avg_ast_score,
        'avg_text_score': avg_text_score,
        'subset_df': subset_df,
        'traceback_error_type': traceback_error_type,
        'subset_size': len(subset_df)
    }

# Example usage:
# For general evaluation (no specific error type)
results = evaluate_diff_metrics(subset_size=100)


```
